```python
%matplotlib widget
%reset -f
%load_ext autoreload
%autoreload 1
%aimport mechanics

from sympy import *
from mechanics import *
from mechanics.lagrange import euler_lagrange_field_equation
import mechanics.config

mechanics.config.diff_notations = {}

BC_0 = 'free' # 'free' or 'fixed'
BC_L = 'fixed'  # 'free' or 'fixed'

t, x = base_spaces('t x')

u, = variables('u', t, x)
rho, kappa, l = constants(r'\rho \kappa \ell')
    
T = (rho / 2) * diff(u, t)**2
U = (kappa / 2) * diff(u, x)**2
L = T - U
show('L = ', L)

EL, BC = euler_lagrange_field_equation(L, (u,))
show_equations(EL)
BC = BC[x][0]
show_equations(BC)
```

```python
%autoreload

from mechanics.difference import forward
from mechanics.space import Z
from mechanics.lagrange import discrete_euler_lagrange_field_equation

i, j = indices('i j')
h, k, T, l = constants(r'h k T \ell')
N, M = constants('N M', space=Z)

d = discretization().space(t, i, h).diff(t, forward)
L_d = (d(L) * h).simplify()
show('L_d =', L_d)

d = discretization().space(x, j, k).diff(x, forward) * d
L_d = (d(L_d) * k).simplify()
show('L_d =', L_d)

EL = discrete_euler_lagrange_field_equation(L_d, (d(u),), (i, 0, N), (j, 0, M))
for eq, indices in EL:
    show_equations(eq.simplify(), indices=indices)
```
